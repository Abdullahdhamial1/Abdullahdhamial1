```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline

print("Downloading real dataset from the internet...")

url = "https://raw.githubusercontent.com/justmarkham/pycon-2016-tutorial/master/data/sms.tsv"
df = pd.read_csv(url, sep='\t', header=None, names=['label', 'text'])

df['label'] = df['label'].map({'ham': 'Safe', 'spam': 'Scam'})

print(f"Data Loaded Successfully! Total Messages: {len(df)}")
print("AI Model training is starting... (Please wait)")

model = make_pipeline(TfidfVectorizer(), SVC(kernel='linear', probability=True))

model.fit(df['text'], df['label'])

print("Model Fully Trained & Ready!\n")
print("-" * 50)

while True:
    user_input = input("Enter a message or link (or press 'q' to quit): ")
    if user_input.lower() == 'q':
        print("System Shutdown.")
        break

    if not user_input.strip():
        continue

    prediction = model.predict([user_input])[0]
    confidence = model.predict_proba([user_input]).max() * 100

    if prediction == "Scam":
        print(f"RESULT: DANGER! This looks like a SCAM! (Confidence: {confidence:.2f}%)\n")
    else:
        print(f"RESULT: SAFE! This message is fine. (Confidence: {confidence:.2f}%)\n")
```
